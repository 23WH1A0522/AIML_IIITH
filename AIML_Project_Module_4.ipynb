```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################
# Define the function and its derivative
def f(x):
    return x**2  # Example function: x^2

def df(x):
    return 2 * x  # Derivative of the function: 2x

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # [REQUIRED] Calculate the gradient
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED] Update x using gradient descent formula

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED] Append the new slope to the list

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
## TASK-7 : Plot reduction in cost for each iteration
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def g(x):
    return x**3 - 3*x**2 + 2  # Example function

def dg(x):
    return 3*x**2 - 6*x  # Derivative of the function

# Gradient Descent Function
def gradient_descent(derivative_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    previous_x = None
    steps = [x]  # Track progress
    costs = [g(x)]  # Track cost at each step

    for _ in range(max_iter):
        previous_x = x
        gradient = derivative_func(previous_x)
        x = previous_x - multiplier * gradient
        steps.append(x)
        costs.append(g(x))

        # Stop if precision is achieved
        if abs(x - previous_x) < precision:
            break

    return {"final_x": x, "steps": steps, "costs": costs}

# Run Gradient Descent for Different Learning Rates
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)
mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n)
high_gamma = gradient_descent(derivative_func=dg, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n)
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=n)

# Plot Cost Reduction Over Iterations
fig, ax = plt.subplots(figsize=(10, 6))

# Plot for each learning rate
ax.plot(low_gamma["costs"], label="Low gamma (0.0005)", color="green")
ax.plot(mid_gamma["costs"], label="Mid gamma (0.001)", color="orange")
ax.plot(high_gamma["costs"], label="High gamma (0.002)", color="red")
ax.plot(insane_gamma["costs"], label="Insane gamma (0.25)", color="purple", linestyle="--")

# Titles and labels
ax.set_title("Reduction in Cost (g(x)) Over Iterations")
ax.set_xlabel("Iteration")
ax.set_ylabel("Cost (g(x))")
ax.legend()

plt.tight_layout()
plt.show()
```

## Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return 1 / (3**(-x**2 - y**2) + 1)

# Partial derivatives
def df_dx(x, y):
    exp_term = 3**(-x**2 - y**2)
    return (2 * x * np.log(3) * exp_term) / (exp_term + 1)**2

def df_dy(x, y):
    exp_term = 3**(-x**2 - y**2)
    return (2 * y * np.log(3) * exp_term) / (exp_term + 1)**2

# Gradient Descent Function for 2D
def gradient_descent_2d(grad_x, grad_y, initial_guess, learning_rate, precision, max_iter):
    x, y = initial_guess
    path = [(x, y)]  # To store steps

    for _ in range(max_iter):
        # Compute gradients
        grad_x_val = grad_x(x, y)
        grad_y_val = grad_y(x, y)

        # Update x and y
        new_x = x - learning_rate * grad_x_val
        new_y = y - learning_rate * grad_y_val

        # Check for convergence
        if np.sqrt((new_x - x)**2 + (new_y - y)**2) < precision:
            break

        x, y = new_x, new_y
        path.append((x, y))

    return {"final_point": (x, y), "path": path}

# Gradient Descent Parameters
initial_guess = (2, 2)  # Starting point
learning_rate = 0.1
precision = 1e-6
max_iter = 100

# Run Gradient Descent
result = gradient_descent_2d(df_dx, df_dy, initial_guess, learning_rate, precision, max_iter)
path = np.array(result["path"])

# Create a grid for visualization
x_vals = np.linspace(-3, 3, 100)
y_vals = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# 3D Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="viridis", alpha=0.8)

# Plot Gradient Descent Path
path_z = f(path[:, 0], path[:, 1])
ax.plot(path[:, 0], path[:, 1], path_z, color="red", marker="o", label="Gradient Descent Path")

# Labels and Title
ax.set_title("Gradient Descent on $f(x, y)$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.legend()

plt.show()
```

```python
# Setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0]) # initial guess

for n in range(max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs={a:params[0], b:params[1]})
    gradient_y = diff(f(a, b), b).evalf(subs={a:params[0], b:params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients

# Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', f(params[0], params[1]))
```

```python

# Setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])  # Initial guess

# Gradient descent loop
for n in range(max_iter):
    # Evaluate the symbolic derivatives numerically at current x and y
    gradient_x = fpx(x, y).subs({x: params[0], y: params[1]}).evalf()  # Numerical value of df/dx
    gradient_y = fpy(x, y).subs({x: params[0], y: params[1]}).evalf()  # Numerical value of df/dy

    gradients = np.array([gradient_x, gradient_y])

    # Update the parameters
    params = params - multiplier * gradients

# Results
print('Values in gradient array:', gradients)
print('Minimum occurs at x value of:', params[0])
print('Minimum occurs at y value of:', params[1])
print('The cost is:', f(params[0], params[1]))
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
import numpy as np

# Gradient function for linear regression
def grad(x, y, thetas):
    n = y.size

    # Calculate partial derivatives (slopes)
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    # Return the slopes as a numpy array
    return np.array([theta0_slope, theta1_slope])

# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
thetas = np.array([0.1, 1.2])  # Initial guess for theta0 and theta1

# Call the gradient function
gradients = grad(x, y, thetas)
print("Gradients:", gradients)
```

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    return np.array([theta0_slope, theta1_slope])

# Initial setup
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

# Run gradient descent for 1000 iterations
for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(mse_vals, mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
